```python
df = pd.DataFrame(np.array([[ 1, 2, 3 ],
     [ 5,  6, 7 ],
     [ 8, 9, 10] ]))
df = np.array(data)
```

```python
sq = np.square(np.array(df)-np.mean(df))
sum = np.sum(sq)
homogeinity_index = np.sqrt(sum/(df.size -1)) / np.sqrt(np.mean(df))
print("H.I. = ",homogeinity_index)
```
